```python
from sklearn import datasets
iris = datasets.load_iris()

x = pd.DataFrame(iris['data'], columns = iris['feature_names'])
y = pd.DataFrame(iris['target'], columns = ['target_names'])
iris_data = pd.concat([x, y], axis = 1)
iris_data.plot.scatter(x = 'sepal length (cm)', y = 'sepal width (cm)', c = 'target_names')
```

```python
color = {
    0:'r',
    1:'g',
    2:'b'
}
iris_data['color'] = iris_data['target_names'].map(color)
iris_data.head(5)
```

```python
iris_data.plot.scatter(x = 'sepal length (cm)', y = 'sepal width (cm)', c = iris_data['color'])
```

```python
%%capture
import plotly.plotly as py
import plotly.graph_objs as go

# Create random data with numpy
import numpy as np

N = 1000
random_x = np.random.randn(N)
random_y = np.random.randn(N)

# Create a trace
trace = go.Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers'
)

data = [trace]

# Plot and embed in ipython notebook!
plotly.offline.iplot(data, filename='basic-scatter')
# or plot with: plot_url = py.plot(data, filename='basic-line')
```

```python
%%capture
N = 100
random_x = np.linspace(0, 1, N)
random_y0 = np.random.randn(N)+5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N)-5

# Create traces
trace0 = go.Scatter(
    x = random_x,
    y = random_y0,
    mode = 'markers',
    name = 'markers'
)
trace1 = go.Scatter(
    x = random_x,
    y = random_y1,
    mode = 'lines+markers',
    name = 'lines+markers'
)
trace2 = go.Scatter(
    x = random_x,
    y = random_y2,
    mode = 'lines',
    name = 'lines'
)

data = [trace0, trace1, trace2]
plotly.offline.iplot(data, filename='scatter-mode')
```
