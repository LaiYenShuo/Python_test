## 第十三天 常用屬性或方法（2）ndarray

```python
import numpy as np
import pandas as pd
ironmen = [56, 8, 19, 14, 6, 71]
ironmen_array = np.array(ironmen)
#print(ironmen)
print(ironmen_array)
print(ironmen_array.ndim) # num of dimensions
print(ironmen_array.shape) # m*n
print(ironmen_array.dtype) # 資料類型
print("\n")

#2d array
ironmen_2d = [range(1, 7), [58, 8, 19, 14, 6, 71]]
ironmen_2d_array = np.array(ironmen_2d)
print(ironmen_2d_array.ndim)
print(ironmen_2d_array.shape)
print(ironmen_2d_array.dtype)

print("---建立 ndarray---")
print(np.zeros(6))
print("------")
print(np.zeros((2, 6)))
print("------")
print(np.empty((2, 6, 2))) # 二十四個元素均為未初始化的值
print("------")
print(np.arange(11)) # 十一個元素為 0 到 10 的 1d array
```

```python
# 轉換變數類型
# ndarray 的 astype() 方法可以轉換變數類型。
import numpy as np
ironmen = ["56", "8", "19", "14", "6", "71"]
ironmen_str_array = np.array(ironmen)
print(ironmen_str_array)
print("---")
# transform int64
ironmen_int_array = ironmen_str_array.astype(np.int64)
print(ironmen_int_array)
# print(ironmen_int_array.dtype)

print("用索引值進行篩選")
# 用索引值進行篩選
# 利用 [] 搭配索引值篩選 ndarray，這點與 R 語言作法相同。
my_array = np.arange(10)
print(my_array[0])
print(my_array[0:5])
print("---")

my_2d_array = np.array([np.arange(0, 5), np.arange(5, 10)])
print(my_2d_array)
print("---")
print(my_2d_array[1, :])
print(my_2d_array[:, 1])
print(my_2d_array[0, 1])

print("用布林值進行篩選")
# 用布林值進行篩選
# 利用布林值（bool）篩選 ndarray，，這點與 R 語言作法相同。

ironmen = [56, 8, 19, 14, 6, 71]
groups = ["Modern Web", "DevOps", "Cloud", "Big Data", "Security", "myself"]
ironmen_array = np.array(ironmen)
groups_array = np.array(groups)
# 用人數去篩選組別
print(ironmen_array >= 10) # 布林值陣列
print(groups_array[ironmen_array >= 10]) # 鐵人數大於 10 的組別

# 用組別去篩選人數
print(groups_array != "myself") 
print(ironmen_array[groups_array != "myself"])
```

```python
# 轉置矩陣
import numpy as np
data = np.arange(12)
data = data.reshape(3, 4)
print(data)
print("---")
print(data.T)

print("numpy 的 where 方法")
# numpy 的 where 方法
# 透過 numpy 的 where() 方法在 ndarray 中進行流程控制。
ironmen_array = np.array([56, 8, 19, 14, 6, np.nan])
print(ironmen_array)
ironmen_array = np.where(np.isnan(ironmen_array), 71, ironmen_array)
print(ironmen_array)
ironmen_array.sort() 
print(ironmen_array)

print("""
---隨機變數---
1. normal dist.
2. uniform dist.
      """)
# 隨機變數
# 透過 numpy 的 random() 方法可以生成隨機變數。
# import random as rd
# help(np.random.normal)
normal_samples = np.random.normal(loc=0, scale=1, size = 10)
print(normal_samples)
#print(normal_samples.mean())
uni_samples = np.random.uniform(size = 10)
print("---")
print(uni_samples)
```

### 爬樹
#### DOM（Document Object Model）的樹狀結構觀念在使用 BeautifulSoup 扮演至關重要的角色，所以我們也要練習爬樹。

### 往下爬
#### 從標籤中回傳更多資訊。

#### $contents 屬性 \\
children 屬性 \\
string 屬性$

### 往上爬
#### 回傳上一階層的標籤。

#### $parent 屬性$

### 往旁邊爬
#### 回傳同一階層的標籤。

#### $next_sibling 屬性 \\
previous_sibling 屬性$



```python
import requests as rq
from bs4 import BeautifulSoup

url = "http://www.ptt.cc/bbs/NBA/index.html" #ptt NBA
response = rq.get(url) # 抓網頁
html_doc = response.text
soup = BeautifulSoup(response.text, "lxml")
print("往下爬")
print(soup.body.a.contents)
print(list(soup.body.a.children))
print(soup.body.a.string)
print("---往上爬---")
print(soup.title)
print("---")
print(soup.title.parent)
print("---往旁邊爬---")
first_a_tag = soup.body.a
next_to_a_tag = first_a_tag.next_sibling
print(first_a_tag)
print("---")
print(next_to_a_tag)
print("---")
print(next_to_a_tag.previous_sibling)

print("---搜尋---")
print(soup.find("a")) # first <a></a>
print("---")
# print(soup.find_all("a")) # all <a></a>
## 加入 class_= 加入CSS的類別

print(soup.find("div", class_= "r-ent"))
```

## BeautifulSoup 牛刀小試
### 把 PTT NBA 版首頁資訊包含推文數，作者 id，文章標題與發文日期搜集下來。

```python
import numpy as np
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup

url = "https://www.ptt.cc/bbs/NBA/index.html"
response = rq.get(url)
html_doc = response.text
soup = BeautifulSoup(html_doc, "lxml")

# posts = soup.find_all("div", class_="r-ent")
#print(posts)
# type(posts)

author_ids = [] # null list id
recommends = [] # null list 推文數
post_title = [] # null list 標題
post_dates = [] # null list Po文日期

posts = soup.find_all("div", class_="r-ent")
for post in posts:
    try:
        author_ids.append(post.find("div", class_="author").string)
    except:
        author_ids.append(np.nan)
        
    try:
        post_title.append(post.find("a").string)
    except:
        post_title.append(np.nan)
        
    try:
        post_dates.append(post.find("div", class_ = "date").string)
    except:
        post_dates.append(np.nan)

# 推文數藏在 div 裡面的 span 所以分開處理
recommendations = soup.find_all("div", class_ = "nrec")
#print(recommendations)
for recommendation in recommendations:
    try:
        recommends.append(int(recommendation.find("span").string))
    except:
        recommends.append(np.nan)
        
#print(author_ids)
#print("---")
#print(post_title)
#print("---")
#print(post_dates)
#print("---")
#print(recommends)

ptt_nba_dict = {
    "author": author_ids,
    "title": post_title,
    "date": post_dates,
    "recommend": recommends
}

ptt_nba_df = pd.DataFrame(ptt_nba_dict)
ptt_nba_df
```
